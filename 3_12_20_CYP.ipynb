```
#Ejemplo 3
email = False
for i in "ariel@.unam.mx":
  if(i == "@"):
    email == True
    
if (email == True):
  print("El email es correcto")
else:
  print("El correo es incorrecto") 
```

```
email = False
miemail= input("Introduce tu correo electronico: ")
for i in miemail:
  if(i == "@"):
    email = True
    
if (email == True):
  print("El correo es correcto")
else:
  print("El correo es incorrecto") 
```

```
edad= int(input("Introduce tu edad "))
while edad < 0:
  print("Has introducido una edad negativa, intentalo de nuevo")
  edad= int(input("Introduce tu edad: "))
print("Gracias por participar, puede pasar")
print(f"Edad del aspirante "+ str(edad))
```

```
edad= int(input("Introduce tu edad "))
while edad < 0 or edad >= 100:
  print("Has introducido una edad negativa, intentalo de nuevo")
  edad= int(input("Introduce tu edad: "))
print("Gracias por participar, puede pasar")
print(f"Edad del aspirante "+ str(edad))
```
